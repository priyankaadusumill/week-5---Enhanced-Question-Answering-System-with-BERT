```python
from transformers import pipeline
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
print("QA pipeline loaded successfully!")
```

```python
context_text = """Big data analytics involves examining large data sets to uncover hidden patterns, correlations, market trends, and other useful information."""
question_text = "What does big data analytics involve?"

result = qa_pipeline({
    'question': question_text,
    'context': context_text
})

print("Context:\n", context_text)
print("\nQuestion:", question_text)
print("\nAnswer:", result['answer'])
print("Confidence Score:", result['score'])
```

```python
%%writefile app.py

import streamlit as st
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="deepset/bert-base-cased-squad2")

st.title("Question Answering System with BERT")
st.write("Enter a context and a question below, and BERT will try to answer it!")

context = st.text_area("Enter the context:", height=200)
question = st.text_input("Enter your question:")

if st.button("Get Answer"):
    if context and question:
        result = qa_pipeline({"context": context, "question": question})
        st.subheader("Answer:")
        st.write(result["answer"])
    else:
        st.warning("Please enter both a context and a question.")
```
